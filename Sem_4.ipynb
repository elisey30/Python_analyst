```python
import pandas as pd
df = pd.read_csv("laptop_price.csv", sep= ',',encoding='windows-1252')
```

```python
df.head()
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(10,4))
plt.hist(df['Memory'])
plt.title('График')
plt.xlabel('GB')
plt.ylabel('Колличество')
plt.xticks(rotation=-90)
```

```python
df['Memory_Type'] = df['Memory'].apply(lambda x: x.split(" ")[1]) 
df.head()
```

```python
def memory (memory_str):
    gb = memory_str.split(" ")[0]
    if gb.endswith('GB'):
        gb = float(gb[:-2])
    elif gb.endswith('TB'):
        gb = float(gb[:-2])*1024
    return gb

df['Memory_Amoun'] = df['Memory'].apply(memory) # 1.2
df.head()
```

```python
plt.figure(figsize=(10,4))
plt.hist(df['Memory_Amoun'])
plt.title('График')
plt.xlabel('GB')
plt.ylabel('Колличество')
plt.xticks(rotation=-90)
```

```python
plt.figure(figsize=(10,4))
plt.hist(df['Weight'])
plt.title('График')
plt.xlabel('вес')
plt.ylabel('Колличество')
plt.xticks(rotation=-90);
```

```python
data5 = pd.crosstab(index=df['Company'], columns=df['Cpy_company'])
data5 = data5.drop(columns=['Samsung']).reset_index()
data5
```

```python
import numpy as np 
```

```python
n_ticks = np.arange(len(data5['AMD']))
offset = 0.2
w = 0.4
plt.bar(n_ticks-offset,data5['AMD'], width=w)
plt.bar(n_ticks+offset,data5['Intel'], width=w)
plt.title('График распределения ноутбуков')
plt.xlabel('Компании')
plt.ylabel('Колличество')
plt.legend(['AMD','Intel'])
plt.xticks(n_ticks,data5['Company'], rotation=90);
```
